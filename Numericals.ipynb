```python
#exact solution
#sample number
N = 1000

#number between 0 and 6.28318531
t = np.linspace(0, 2*pi, N)

#difference between final and initial succesively in the list
h = t[1] - t[0]

#initial values 
x_initial = 1
y_initial = 0
```

```python
def Euler_system2(x_initial,y_initial,t_initial, t_final, n):
    h = (t_final-t_initial) / (n-1)
    x = np.zeros([n])
    x[0] = x_initial
    y = np.zeros([n])
    y[0] = y_initial

    for i in range(1, n):
        x[i] = x[i-1] + h*y[i-1]
        y[i] = -h*x[i-1] + y[i-1]
    plt.plot(x,y, 'b--')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Trajectories for instable fix point")
    plt.show()

Euler_system2(1,0,0,2*pi,1000)
Euler_system2(1,0,0,4*pi,1000)
Euler_system2(1,0,0,16*pi,1000)
Euler_system2(1,0,0,32*pi,1000)
```
