```python
#importing libraries to be used
import numpy as np
import matplotlib.pyplot as plt
from math import pi
%matplotlib inline
import warnings
warnings.filterwarnings("ignore")
print('Start')
```

```python
# Numerical solution and values
#sample number
N = 1000

#number between 0 and 6.28318531
t = np.linspace(0, 2*pi, N)

#difference between final and initial at every successive step in the list of number
h = t[1] - t[0]

#initial values 
x_initial = 1
y_initial = 0
```

```python
# In this code, we will again see the graph for the exact solution
# and also compare it with the approximate solution. 
x = np.cos(t)
y = np.sin(-t)
plt.plot(x, y, 'y')
plt.xlabel('x(t)')
plt.ylabel('y(t)')
plt.title('The Exact solution')
plt.show()
```

```python
#  Define a function called the semi-implicit Euler method
def semi_implicit_euler(x_initial,y_initial,t_initial, t_final, n):
    h = (t_final-t_initial) / (n-1)
    x = np.zeros([n])
    x[0] = x_initial
    y = np.zeros([n])
    y[0] = y_initial

    # Loop  
    for i in range(1, n):
        x[i] = x[i-1] + h*y[i-1]
        y[i] = -h*x[i-1] + y[i-1]
    plt.plot(x,y, 'b--')
    plt.xlabel("x")
    plt.ylabel("y")
    plt.title("Trajectory ")
    plt.show()
    
# we call the function 'semi_implicit_euler' and visalized  
semi_implicit_euler(1,0,0,2*pi,1000)
semi_implicit_euler(1,0,0,4*pi,1000)
semi_implicit_euler(1,0,0,16*pi,1000)
semi_implicit_euler(1,0,0,32*pi,1000)
```
